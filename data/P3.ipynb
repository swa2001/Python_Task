# P3> Find the best Rooms under each Type in each city based on Reviews


```python
import pandas as pd
```

```python
df = pd.read_excel("/home/nineleaps/Python_training/data/Sales_Data/cleaned_data.xlsx")
df.head()
```

### Finding best rooms for each city ,Ptype by sorting and grouby

```python
best_rooms = (
    df.sort_values(['City', 'PType', 'Reviews'], ascending=[True, True, False])  
    .groupby(['City', 'PType'])  
    .head(1) 
)

best_rooms 
```
