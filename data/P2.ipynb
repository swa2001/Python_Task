# P2: Find the Top 3 Rooms in Each City Based on Reviews

```python
import pandas as pd
```

```python
df = pd.read_excel("/home/nineleaps/Python_training/data/Sales_Data/cleaned_data.xlsx")
df.head()
```

### finding top 3 rooms by sorting and grouping it by city

```python
top3_rooms = (
    df.sort_values(['City', 'Reviews'], ascending=[True, False])  
    .groupby('City', group_keys=False) 
)

top3_rooms 
```
