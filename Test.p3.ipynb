```python
import pandas as pd


bollywood_df = pd.read_excel("bollywood.xlsx")
tollywood_df = pd.read_excel("tollywood.xlsx")
indian_movies_df = pd.read_excel("indian.xlsx")

```

```python
movies_df = pd.concat([bollywood_df, tollywood_df, indian_movies_df], ignore_index=True)
```

```python
# getting all the bollywood movies 

bollywood_movies = movies_df[movies_df["Language"] == "Hindi"]
```

```python
# Grouping by Genre and calculating average Box Office collection

bollywood_genre_avg = bollywood_movies.groupby("Genre")["BoxOffice (Crores)"].mean().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
```

```python
# Plot the bar chart
plt.figure(figsize=(8, 6))
plt.barh( bollywood_genre_avg.index,  bollywood_genre_avg.values, color="green", alpha=0.7)
plt.ylabel("Genre")  # Now on the Y-axis
plt.xlabel("Average Box Office Collection (Crores)")  # Now on the X-axis
plt.title("Average Box Office Collection by Genre (All Movies)")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```
