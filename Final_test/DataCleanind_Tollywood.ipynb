Cleaning Tollywood Dataset

```python
import pandas as pd
```

```python
df=pd.read_excel('/home/nineleaps/Python_training/Trainer_Assignment/tollywood_movies.xlsx')
df
```

```python
df.rename(columns={'Unnamed: 0': 'index'}, inplace=True)
df.head()
```

```python
df.info()
```

```python
df[df['ReleaseYear'].isna()]
```

```python
df.loc[df["MovieID"] == "MOV004", "ReleaseYear"] = 2017
```

```python
df.loc[df["MovieID"] == "MOV034", "ReleaseYear"] = 2020
```

```python
df['ReleaseYear']=df['ReleaseYear'].astype(int)
```

```python
df.info()
```

```python
df['Budget (Crores)']=df['Budget (Crores)'].astype(float)
df['BoxOffice (Crores)']=df['BoxOffice (Crores)'].astype(float)
```

```python
df.info()
```

```python
df[df['Rating'].isna()]
```

```python
df['Rating']=df['Rating'].fillna(7.4)
df.info()
```

```python
df[df['Duration (minutes)'].isna()]
```

```python
df['Duration (minutes)']=df['Duration (minutes)'].fillna(128).astype(int)
df.info()
```

```python
df[df['Language'].isna()]
```

```python
df['Language']=df['Language'].fillna('Telugu')
df.info()
```

```python
df[df['ProductionCompany'].isna()]
```

```python
df['ProductionCompany']=df['ProductionCompany'].fillna('Arka Media Works')
df.info()
```

```python
df.to_excel('tollywood.xlsx',index= False)
```
