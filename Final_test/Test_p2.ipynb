P2> Plot the distribution of BoxOffice collections in the combined dataset
What is your inference?
Are there any outliers? Who are they?


we will plot the histogram and boxplot to see the distribution and if any outliers present


```python
import pandas as pd


bollywood_df = pd.read_excel("bollywood.xlsx")
tollywood_df = pd.read_excel("tollywood.xlsx")
indian_movies_df = pd.read_excel("indian.xlsx")


```

```python
movies_df = pd.concat([bollywood_df, tollywood_df, indian_movies_df], ignore_index=True)

print(movies_df.head())
```

```python
import matplotlib.pyplot as plt
```

```python
# Histogram to show the distribution
plt.figure(figsize=(8, 5))
plt.hist(movies_df["BoxOffice (Crores)"], bins=30, color="blue", edgecolor="black", alpha=0.7)
plt.xlabel("Box Office Collection (Crores)")
plt.ylabel("Number of Movies")
plt.title("Distribution of Box Office Collections")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

Observations from the Histogram
* The distribution of Box Office collections is skewed, means most movies have lower earnings and few have high collections.
* Most of the movies earn below 500 crores



```python
# Boxplot to detect outliers
plt.figure(figsize=(8, 5))
plt.boxplot(movies_df["BoxOffice (Crores)"], vert=False, patch_artist=True, boxprops=dict(facecolor="red"))
plt.xlabel("Box Office Collection (Crores)")
plt.title("Box Plot of Box Office Collections")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```

Observations from the Box Plot:
* The box plot confirms the presence of outliers as we can see few movies have box office collections that are higher than the rest.
* There are two extreme points beyond 1500 crores are  outliersthese are mostly record-breaking movies with good performance.
