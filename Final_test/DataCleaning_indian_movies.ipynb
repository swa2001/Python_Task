Data cleaning for indian_movies dataset

```python
import pandas as pd
```

```python
df=pd.read_csv('/home/nineleaps/Python_training/Trainer_Assignment/indian_movies.csv')
df.head()
```

```python
df.info()
```

```python
# since the column 1 is unnamed i am naming as index as per previous dataset

df.rename(columns={'Unnamed: 0': 'index'}, inplace=True)
df.head()
```

```python
df[df.ReleaseYear.isna()]
```

```python
# release year is missing and its datatype shd not be float 

# i refered to the other dataset for the release year of the movie U turn and found that it was relesed in 2016 so i am filling it with this year

df['ReleaseYear']=df['ReleaseYear'].fillna(2016)
```

```python
df['ReleaseYear']=df["ReleaseYear"].astype(int)
df.info()
```

```python
# Budget and Boxeffice has 8 null values will fill that with mean

```

```python
df[df['Budget (Crores)'].isna()]
```

```python
df['Budget (Crores)']=df['Budget (Crores)'].fillna(df["Budget (Crores)"].median())
```

```python
df['BoxOffice (Crores)']=df['BoxOffice (Crores)'].fillna(df['BoxOffice (Crores)'].median())
```

```python
df.info()
```

```python
df[df['Rating'].isna()]
```

```python
# i refered to  the bollywood dataset for the title sahoo and filling the null val with its rating here

df['Rating']=df['Rating'].fillna(6.5)

```

```python
df[df['Duration (minutes)'].isna()]
```

```python
df['Duration (minutes)']=df['Duration (minutes)'].fillna(166).astype(int)
df.info()
```

Data is cleaned no missing values and inconsistency


```python
df.to_excel('indian.xlsx',index=False)
```
