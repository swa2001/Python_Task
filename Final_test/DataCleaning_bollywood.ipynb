Data cleaning for bollywood_movies dataset


```python
# import the necessary modules

import sqlite3
import pandas as pd
```

```python
# connecting to the sqlite database

connect = sqlite3.connect("bollywood_movies.sqlite")
```

```python
# getting the list of all the tables present in the database

query = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql(query, connect)
print("Tables in the database:\n", tables)
```

```python
# loading all the data from table into dataframe
query = "SELECT * FROM Movies"
df = pd.read_sql(query, connect)
df.head()
```

```python
# closing the connection as we have loaded data into our dataframe

connect.close()
```

```python
df.info()
```

```python
# checking for missing data in genre column

df[df.Genre.isnull()]
```

```python
df['Genre'].unique()
```

```python
# since only 1 data is missing i can fill the null value acc to the movie

df['Genre'] = df['Genre'].fillna("Sports Drama")
```

```python
df.info()
```

```python
# checking missing values in Budget col

df[df['Budget (Crores)'].isna()]
```

```python
# i can fill the null values with mean

df['Budget (Crores)']=df['Budget (Crores)'].fillna(df["Budget (Crores)"].mean())
df.info()
```

```python
# checking missing values in BoxOffic col

df[df['BoxOffice (Crores)'].isna()]
```

```python
df['BoxOffice (Crores)']=df['BoxOffice (Crores)'].fillna(df['BoxOffice (Crores)'].median())
df.info()
```

```python
# now the data is cleaned 
df
```

```python
df.to_excel('bollywood.xlsx',index=False)
```
