```python
import pandas as pd
```

```python
folder  = "data"
filename = "sales_updated.csv"
path = f'{folder}/{filename}'
path
```

```python
df = pd.read_csv(path)
df.head()
```

```python
df['Region'] = df.Region.replace(
    {
    'EAst' : 'East',
    'SOUTH' : 'South',
    },
regex = True
)
df.head()
```

```python
df['Region'] = df['Region'].fillna('East')
df
```

```python
df['Product'].unique
```

```python
df['Product'] = df.Product.replace(
    {
    'g2-Blazers' : 'G2-Blazers',
    'D1Shirt' : 'D1-Shirt',
    },
regex = True
)
df
```

```python
df.info()
```

```python
# pd.to_datetime(df['Date'], errors='coerce').dt.strftime('%d/%m/%Y')

df['Date'] = pd.to_datetime(df['Date'],errors='ignore',yearfirst = True)
df
```

```python
df.Date = df.Date.replace(
    {
    '15/01/2024' : '2024-01-15',
    '24/2/20' : '2020-02-24',
    '13/3/24' : '2024-03-13',
    '21/04/2024' : '2024-04-21',
    '27/5/24' : '2024-05-27'
    },
regex = True
)
df
```

# P1> Find which month had the highest sales

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df.info()
```

```python
df["Month"] = df["Date"].dt.month
```

```python
month_sales = df.groupby("Month")["Amount"].sum()
```

```python
# Find the month with the highest sales
highest_sales_month = month_sales.idxmax()
highest_sales_value = month_sales.max()
```

```python
print(f"The month with the highest sales is: {highest_sales_month} with sales of {highest_sales_value}")
```

# P2> Plot region-wise sales data showing the highest sales 


```python
# Group by Region and sum sales
region_sales = df.groupby("Region")["Amount"].sum().sort_values(ascending=False)

```

```python
# Plot the data
import matplotlib.pyplot as plt 
plt.figure(figsize=(8, 5))
region_sales.plot(kind="bar", color="skyblue")
plt.title("Region-wise Sales")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

 # P3> Plot month-wise sales data and give an inference

```python
# Plot month-wise sales
plt.figure(figsize=(8, 5))
month_sales.plot(kind="bar", color="lightcoral")
plt.title("Month-wise Sales")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

Inference from the Plot:
June (Month 6) has the highest sales, meaning there was a peak in sales activity.
Sales were strong in May (Month 5) and February (Month 2)
March (Month 3) has the lowest sales, which may indicate a slow period for sales, possibly due to reduced demand, fewer promotions,
