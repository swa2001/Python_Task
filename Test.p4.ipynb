```python
import pandas as pd


bollywood_df = pd.read_excel("bollywood.xlsx")
tollywood_df = pd.read_excel("tollywood.xlsx")
indian_movies_df = pd.read_excel("indian.xlsx")

```

```python
movies_df = pd.concat([bollywood_df, tollywood_df, indian_movies_df], ignore_index=True)
```

 Trend of Box Office Collection Over the Years

```python
yearly_avg = movies_df.groupby("ReleaseYear")["BoxOffice (Crores)"].mean()
```

```python
import matplotlib.pyplot as plt
```

```python

plt.figure(figsize=(10, 5))
plt.plot(yearly_avg.index, yearly_avg.values, marker="o", linestyle="-", color="purple")
plt.xlabel("Release Year")
plt.ylabel("Average Box Office Collection (Crores)")
plt.title("Trend of Box Office Collection Over the Years")
plt.grid(True)
plt.show()
```

Most Successful Production Companies

```python
top_productions = movies_df.groupby("ProductionCompany")["BoxOffice (Crores)"].sum().sort_values(ascending=False).head(10)
```

```python

plt.figure(figsize=(8, 5))
plt.barh(top_productions.index, top_productions.values, color="red", alpha=0.7)
plt.xlabel("Total Box Office Collection (Crores)")
plt.ylabel("Production Company")
plt.title("Top 10 Highest-Grossing Production Companies")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```
